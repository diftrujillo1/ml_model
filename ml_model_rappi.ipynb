## df_data (after merging with other tables)

```python
df_data[df_data.Demanda_uni_equil < 0]
df_data['return'] = df_data.Venta_uni_hoy - df_data.Dev_uni_proxima
# número de veces en donde Venta_uni_hoy < Dev_uni_proxima
len(df_data[df_data['return'] < 0])
```

```python
df_data[df_data.Demanda_uni_equil == 0]
```

```python
def function_desc(df):
    conteo = df.Venta_hoy.count()
    ventas = df.Venta_hoy.sum()
    unidades = df.Venta_uni_hoy.sum()
    dev_unidades_prox = df.Dev_uni_proxima.sum()
    dev_prox = df.Dev_proxima.sum()
    mayor_returns = len(df[df.Demanda_uni_equil == 0])
    dev_unidades_equil = df.Demanda_uni_equil.sum()
    returns = df['return'].sum()
    ids_cls_unique = df.Cliente_ID.unique() 
    ids_prod_unique = df.Producto_ID.unique()
    ids_agen_unique = df.Agencia_ID.unique()
    conteo_clientes = len(ids_cls_unique)
    conteo_agencias = len(ids_agen_unique)
    conteo_prods = len(ids_prod_unique)
    
    
    return pd.Series(dict(
        conteo = conteo,
        ventas = ventas,
        unidades = unidades,
        dev_unidades_prox = dev_unidades_prox,
        dev_prox = dev_prox,
        dev_unidades_equil = dev_unidades_equil,
        returns = returns,
        conteo_clientes = conteo_clientes,
        conteo_agencias = conteo_agencias,
        conteo_prods = conteo_prods,
        ids_cls_unique = ids_cls_unique,
        ids_prod_unique = ids_prod_unique,
        ids_agen_unique = ids_agen_unique,
        mayor_returns = mayor_returns
    ))

# análisis agrupado
df_data.groupby('Semana').apply(function_desc)
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(df_data, test_size=0.2, random_state=50)
```

### Categorización de variables TRAIN test

```python
dummies_state = pd.get_dummies(train.State)
dummies_state = dummies_state.apply(lambda x: x * train.Demanda_uni_equil.values)

dummies_semana = pd.get_dummies(train.Semana, prefix='semana_')
dummies_semana = dummies_semana.apply(lambda x: x * train.Demanda_uni_equil.values)

dummies_canal = pd.get_dummies(train.Canal_ID, prefix='canal_')
dummies_canal = dummies_canal.apply(lambda x: x * train.Demanda_uni_equil.values)

train_cat = pd.concat([train, dummies_state], axis=1)      
train_cat = pd.concat([train_cat, dummies_semana], axis=1)      
train_cat = pd.concat([train_cat, dummies_canal], axis=1)      
train_cat.drop(['State', 'Semana', 'Agencia_ID', 'Canal_ID'], inplace=True, axis=1)
train_cat
```

### Categorización de variables TEST test

```python
dummies_state = pd.get_dummies(test.State)
dummies_state = dummies_state.apply(lambda x: x * test.Demanda_uni_equil.values)

dummies_semana = pd.get_dummies(test.Semana, prefix='semana_')
dummies_semana = dummies_semana.apply(lambda x: x * test.Demanda_uni_equil.values)

dummies_canal = pd.get_dummies(test.Canal_ID, prefix='canal_')
dummies_canal = dummies_canal.apply(lambda x: x * test.Demanda_uni_equil.values)

test_cat = pd.concat([test, dummies_state], axis=1)      
test_cat = pd.concat([test_cat, dummies_semana], axis=1)      
test_cat = pd.concat([test_cat, dummies_canal], axis=1)      
test_cat.drop(['State', 'Semana', 'Agencia_ID', 'Canal_ID'], inplace=True, axis=1)
test_cat
```

# Model

## prueba con producto 1212

```python
cols = list(set(train_cat.columns) - set(['Ruta_SAK', 'Cliente_ID', 'Producto_ID', 
                                            'NombreCliente', 'NombreProducto', 'Town']))

train_cat_filtered = train_cat[train_cat.Producto_ID == prueba_producto][cols]

test_cat_filtered = test_cat[test_cat.Producto_ID == prueba_producto][cols]

train_cat_filtered
```

```python
Y_train = train_cat_filtered[['Demanda_uni_equil']].values
X_train = train_cat_filtered[list(set(cols) - set(['Demanda_uni_equil']))]
X_train = X_train.apply(pd.to_numeric)

Y_test = test_cat_filtered[['Demanda_uni_equil']].values
X_test = test_cat_filtered[list(set(cols) - set(['Demanda_uni_equil']))]
X_test = X_test.apply(pd.to_numeric)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score

regr = DecisionTreeRegressor(max_depth=5)
regr.fit(X_train, Y_train)

y_pred = regr.predict(X_test)
pd.DataFrame({'y_real': np.ravel(Y_test), 'y_pred': np.ravel(y_pred)})
```

## Predicción


```python
print(r2_score(Y_test, y_pred))
```
